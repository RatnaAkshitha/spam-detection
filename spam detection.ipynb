```python
import tensorflow as tf
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
voc_size=50000
```

```python
messages=X.copy()
```

```python
messages.head()
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download("stopwords")
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0, len(messages)):
    
    review = re.sub('[^a-zA-Z]', ' ', messages['v2'][i])
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
onehot_repr=[one_hot(word,voc_size) for word in corpus]
onehot_repr
```

```python
sent_length=50
embedded_docs=pad_sequences(onehot_repr,padding="pre",maxlen=sent_length)
print(embedded_docs)
```

```python
from tensorflow.keras.layers import Bidirectional
## Creating model
embedding_vector_features=40
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add((LSTM(100)))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python
import numpy as np
X_final=np.array(embedded_docs)
y_final=np.array(y)
```

```python
## train test split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final,y_final, test_size=0.33, random_state=42)
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=32)
```

```python
# Save the trained model
model.save('spam_ham_classifier.h5')
from tensorflow.keras.models import load_model

# Load the saved model
loaded_model = load_model('spam_ham_classifier.h5')
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred=np.where(y_pred>=0.5,1,0)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
confusion_matrix(y_test,y_pred)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
# Function to Predict New Messages
def predict_message(message, model, voc_size, sent_length):
    ps = PorterStemmer()
    review = re.sub('[^a-zA-Z]', ' ', message)
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    onehot_repr = [one_hot(review, voc_size)]
    embedded_docs = pad_sequences(onehot_repr, padding="pre", maxlen=sent_length)
    prediction = model.predict(embedded_docs)
    return 'spam' if prediction >= 0.5 else 'ham'

# Example Prediction
user_input = input("Enter a message to check if it's spam or ham: ")
result = predict_message(user_input, model, voc_size, sent_length)
print(f"The message is: {result}")
```
